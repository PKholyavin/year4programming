Фильтры scipy

```python
import numpy as np

M = 1024

signal = np.sin(np.arange(M) / 10) + np.sin(np.arange(M) + 10) + np.sin(np.arange(M) * 10 - 6)

f1 = M / (2 * np.pi) / 20
f2 = M / (2 * np.pi) / 10
f3 = M / (2 * np.pi) * 2

print(f1, f2, f3)
```

```python
import matplotlib.pyplot as plt
plt.plot(signal)
plt.show()
```

```python
from scipy.fft import fft
spectrum = np.abs(fft(signal))
plt.plot(spectrum[:M // 2])
```

```python
from scipy.signal import butter, lfilter

low_pass_filter = butter(5, 200 / (M / 2))

signal_filtered = lfilter(*low_pass_filter, signal)
```

```python
plt.plot(signal_filtered)
plt.show()
```

```python
from scipy.fft import fft
spectrum_filtered = np.abs(fft(signal_filtered))
plt.plot(spectrum_filtered[:M // 2])
```
