Функция автокорреляции

```python
import numpy as np
import matplotlib.pyplot as plt
d = np.sin(np.arange(0, np.pi * 10, np.pi / 20)) + np.sin(1.5 * np.arange(0, np.pi * 10, np.pi / 20))
plt.plot(d[:100])
```

```python
frame = 100
autocorr = [sum(d[i] * d[i + lag] for i in range(frame)) for lag in range(frame)]

plt.plot(autocorr)
plt.show()
```
