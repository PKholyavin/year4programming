Для оценки значения частоты основного тона можно использовать функцию автокорреляции, которая показывает, в какие моменты сигнал начинает повторять сам себя.

Для непрерывного сигнала:

$$ r_x(\tau) = \int x(t) x(t + \tau) dt$$

Для дискретного сигнала:

$$ r_x[\tau] = \sum_{t=0}^{N-1}x[t] \cdot x[t + \tau]$$

где $\tau$ &ndash; задержка, $N$ &ndash; размер окна.

![](https://pkholyavin.github.io/year4programming/autocorr.png)

Автокорреляцию принято нормализовывать:

$$r'_x[\tau] = \frac{r_x[\tau]}{r_x[0]}$$

Максимум функции автокорреляции покажет, с каким сдвигом сигнал лучше всего накладывается сам на себя.
